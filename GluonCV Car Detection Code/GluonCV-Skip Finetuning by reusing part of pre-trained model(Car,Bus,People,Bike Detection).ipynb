```python
from matplotlib import pyplot as plt
import gluoncv
from gluoncv import model_zoo, data, utils
```

```python
net = model_zoo.get_model('faster_rcnn_resnet50_v1b_coco', pretrained=True)
```

```python
net2 = model_zoo.get_model('ssd_512_resnet50_v1_voc', pretrained=True)
```

```python
net3 = model_zoo.get_model('yolo3_darknet53_voc', pretrained=True)
```

```python
net4 = model_zoo.get_model('ssd_300_vgg16_atrous_voc', pretrained=True)
```

```python
net5 = model_zoo.get_model('mask_rcnn_resnet50_v1b_coco', pretrained=True)
```

```python
x, orig_img = data.transforms.presets.rcnn.load_test("img00830.jpg")
```

```python
print('coco classes: ', net.classes)
net.reset_class(classes=["bicycel","car","person","motorbike","bus"], 
                reuse_weights=["bicycel","car","person","motorbike","bus"])
print('new classes: ', net.classes)

box_ids, scores, bboxes = net(x)
ax = utils.viz.plot_bbox(orig_img, bboxes[0], scores[0], box_ids[0], class_names=net.classes)

plt.savefig("img00830.png")
plt.show()
```

```python
print('voc classes: ', net2.classes)
net2.reset_class(classes=["bicycel","car","person","motorbike","bus"], 
                reuse_weights=["bicycel","car","person","motorbike","bus"])
print('new classes: ', net2.classes)
x2, orig_img = data.transforms.presets.rcnn.load_test("img00830.jpg")
box_ids, scores, bboxes = net2(x2)
ax = utils.viz.plot_bbox(orig_img, bboxes[0], scores[0], 
                         box_ids[0], class_names=net2.classes)
plt.savefig("img00830-2.png")
plt.show()
```
